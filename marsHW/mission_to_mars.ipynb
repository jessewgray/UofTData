```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

```python
import time
import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

chromedriver = "/usr/local/bin/chromedriver"
os.environ["webdriver.chrome.driver"] = chromedriver
driver = webdriver.Chrome(chromedriver)


from splinter import Browser
executable_path = {'executable_path':'/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path)
```

```python
marsNewsUrl = 'https://mars.nasa.gov/news/'
browser.visit(marsNewsUrl)
marsNewsButton = browser.find_by_css('a[class="button"]')

for i in range(0,2):
    marsNewsButton.click()
    time.sleep(8)
```

```python
marsNewsPage = browser.html
soup = bs(marsNewsPage, 'lxml')

titleAndTextList = []
listText = soup.find_all('div', class_='list_text')
for list in listText:
    contentTitleDiv = list.find('div', class_='content_title')
    theTitle = contentTitleDiv.a.text.strip()
    teaserDiv = list.find('div', class_='article_teaser_body')
    teaserText = teaserDiv.text.strip()
    article = {"title": theTitle, "teaser": teaserText}
    titleAndTextList.append(article)
titleAndTextList
```

```python
theTDS = []
response = requests.get('http://space-facts.com/mars/')
soup = bs(response.text, 'lxml')
table = soup.table
table

```

```python
theTR = table.tbody.find_all('tr')
for row in theTR:
    tds = row.find_all('td')
    for td in tds:
        #print(td.text.strip())
        theTDS.append(td.text.strip())
theTDS
```

```python
colOne = []
colTwo = []
for i in range(len(theTDS)):
    if i == 0:
        colOne.append(theTDS[i])
    elif i % 2:
        colTwo.append(theTDS[i])
    else:
        colOne.append(theTDS[i])
colOne
```

```python
dfObj = {'colOne': colOne, 'colTwo': colTwo}
df = pd.DataFrame(dfObj)
df
```

```python
marsAstro = "https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars"
browser.visit(marsAstro)

marsAstroPage = browser.html
soup = bs(marsAstroPage, 'lxml')

soup

```

```python
marsArray = []
baseHrefs = 'https://astrogeology.usgs.gov'
itemLinks = soup.find_all('div', class_='item')
for item in itemLinks:
    theTitle = item.find('h3')
    theTitle = theTitle.text
    theURL = baseHrefs + item.a.attrs['href']
    browser.visit(theURL)
    clickToPage = browser.html
    newSoup = bs(clickToPage, 'lxml')
    wideImage = newSoup.find('img', class_='wide-image')
    wideImageSrc = baseHrefs + wideImage.attrs['src']
    marsArray.append({'title':theTitle, 'imageUrl':wideImageSrc})
marsArray
    
```
